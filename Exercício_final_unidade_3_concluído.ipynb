```python
import sqlite3

# Conectando ao banco de dados
conn = sqlite3.connect('dados_vendas.db')

# Criando cursor
cursor = conn.cursor()

# Executando uma consulta para obter todos os dados da tabela Vendas1
cursor.execute("SELECT * FROM Vendas1")

# Buscando todos os resultados da consulta
dados_vendas = cursor.fetchall()

# Imprimindo os dados das vendas
print(dados_vendas)

for dados_venda in dados_vendas:
    print(dados_venda)
#Fechando conexão
conn.close()
```

```python
import pandas as pd

df_vendas = pd.DataFrame(dados_vendas, columns=['id_venda', 'data_venda', 'produto', 'categoria', 'valor_venda'])

```

```python
# Grafico com os produtos com melhores valores em vendas
df_vendas.plot(x='produto', y='valor_venda', kind='bar')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Gráfico com os valores de vendas individuais, por produto
sns.set(style="darkgrid")

sns.barplot(x='produto', y='valor_venda', data=df_vendas)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
print(df_vendas) #Visualização da nossa biblioteca
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

#Gráfico com a média gasta por categoria
sns.barplot(data=df_vendas, x='categoria', y='valor_venda', ax=ax[0])


#Grafico classificando as categorias por fauramento
sns.barplot(data=df_vendas, x='categoria', y='valor_venda', ax=ax[1], estimator=sum)


#Gráfico demonstrando quantos produtos foram vendidos por categoria
sns.barplot(data=df_vendas, x='categoria', y='valor_venda', ax=ax[2], estimator=len)

plt.tight_layout()
plt.show()
```

```python
#Grafico apresentando o ticket médio de cada categoria de produto vendido na empresa

plt.figure(figsize=(8,5))
sns.barplot(x='categoria', y='valor_venda', data=df_vendas) #estimator=sum, ci=None, palette="Set2")
plt.xlabel('Categoria do produto')
plt.ylabel('valor do produto')
plt.title('Ticket médio de de venda por categoria')
plt.show()
```

```python
#Grafico com o valor total de vendas por categoria
plt.figure(figsize=(8,5))
sns.barplot(x='categoria', y='valor_venda', data=df_vendas, estimator=sum)
plt.xlabel('Categoria')
plt.ylabel('Total de vendas')
plt.title('Valor total de vendas por categoria')
plt.show()
```

```python
#Grafico a quantidade de produtos vendidos por categoria
plt.figure(figsize=(8,5))
sns.barplot(x='categoria', y='valor_venda', data=df_vendas, estimator=len)
plt.xlabel('Categoria')
plt.ylabel('Quantidade')
plt.title('Quantidade de produtos vendidos por categoria')
plt.show()
```
