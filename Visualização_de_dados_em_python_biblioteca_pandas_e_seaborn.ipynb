```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid") #Opes darkgrid, whitegrid, dark, white, ticks

df_tips = sns.load_dataset('tips')
print(df_tips)

```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

#Media por sexo
sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[0])

#O motivo para optar pelo 'barplot()' muitas veze se baseia nos parametros adicionais e na flexibilidade que ele oferece vamos dar destaque ao parametro "estimator", que calcula a media
sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[1], estimator=sum)


#A funo "barplot()" do seaborn apresenta uma variedade de opes estatisticas permitindo que os cientistas de dados escolham a metrica que melhor se ajudsta a seus objetivos
#Por exemplo, voce pode calcular a soma, a contagem e at mesmo outras metricas personalizadas isso  particularmente til quando voce deseja exibir informaes diferentes nas barras
#como a quantidade (len) ou a soma (sum) dos valores ao invs da mdia

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[2], estimator=len)
```

```python
#Exercicio : Precisamos responder qual o periodo os clientes gastam mais em um restaurante e se esse mesmo periodo  o que eles dao mais gorjetas, mostre com graficos

import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid") #Opes darkgrid, whitegrid, dark, white, ticks

df = sns.load_dataset('tips')

plt.figure(figsize=(8,5))
sns.barplot(x='time', y='total_bill', data=df, estimator=sum, ci=None, palette="Set2")
plt.xlabel('Perodo (Time)')
plt.ylabel('Total de Gastos')
plt.title('Total de Gastos por perodo (Almoo ou jantar)')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='time', y='total_bill', data=df) #estimator=sum, ci=None, palette="Set2")
plt.xlabel('Perodo (Time)')
plt.ylabel('Mdia de Gastos')
plt.title('Mdia de Gastos por perodo (Almoo ou jantar)')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='time', y='total_bill', data=df, palette="Set3")
plt.xlabel('Perodo (Time)')
plt.ylabel('Mdia de Gorjeta')
plt.title('Mdia de Gorjeta por perodo (Almoo ou jantar)')
plt.show()
```
